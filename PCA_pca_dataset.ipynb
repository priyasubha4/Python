```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
uni = pd.read_csv("/content/PCA.csv")
```

```python
# We are getting a categorical data
print(uni.describe())
uni.head() # output of 1st five rows
```

```python
# Converting into numerical data as statistics work only with numbers
uni.data = uni.iloc[:,1:]
uni.data.head()
# decimal removed, getting a uniform data
```

```python
# Converting into numpy arrays
UNI = uni.data.values
UNI
```

```python
uni.data
```

```python
# Normalizing the numerical data
from sklearn.preprocessing import scale
uni_normal = scale(UNI)
```

```python
uni_normal
```

```python
#fitting the data Model
from sklearn.decomposition import PCA
pca = PCA()
pca_values = pca.fit_transform(uni_normal)
```

```python
pca_values
```

```python
# The amount of variance that each PCA explains is
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative Variance
var1 = np.cumsum(np.round(var, decimals = 4)* 100)
var1
```

```python
# Variance plot for PCA components obtained
plt.plot(var1, color ="red")# Elbow plot like K-means 
```

```python
pca.components_
```

```python
pca_values[:, 0:2]
```

```python
# plot between PCA1 and PCA2
x = pca_values[:, 0:1]
y = pca_values[:, 1:2]
plt.scatter(x,y)
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values [:, 0:2], columns = ['pc1','pc2']),  uni[['Univ']]], axis =1)
finalDf
```

```python
import seaborn as sns
sns.scatterplot(data = finalDf, x='pc1', y='pc2', hue = 'Univ')
plt.legend(loc = 'upper left', bbox_to_anchor =(1,1))
```
